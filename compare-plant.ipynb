# Comparison by hardware and frequency

This analysis shows the FRF plant for three different hardware measuring a generic bicycle hub motor.
All three show good agreement at 10 kHz with each other and expected values.
One of the hardware shows a change in the frequency response as the switching frequency increases.

```python
%load_ext autoreload
%autoreload 2
```

```python
import matplotlib.pyplot as plt
import vesc_frf as vf

data_directories = {'10 kHz': ('ENNOID-FRF/Q-V-RL5A-10kHz-9CTB/',
                               'FSESC-420-FRF/Q-V-RL5A-10kHz/',
                               'FSESC-7550-FRF/Q-V-RL5A-10kHz/'),
                    '20 kHz': ('ENNOID-FRF/Q-V-RL5A-20kHz-9CTB/',
                               'FSESC-420-FRF/Q-V-RL5A-20kHz/',
                               'FSESC-7550-FRF/Q-V-RL5A-20kHz/'),
                    '30 kHz': ('ENNOID-FRF/Q-V-RL5A-30kHz-9CTB/',
                               'FSESC-420-FRF/Q-V-RL5A-30kHz/',
                               'FSESC-7550-FRF/Q-V-RL5A-30kHz/'),
                    '40 kHz': ('ENNOID-FRF/Q-V-RL5A-40kHz-9CTB/',
                               'FSESC-420-FRF/Q-V-RL5A-40kHz/',
                               'FSESC-7550-FRF/Q-V-RL5A-40kHz/'),
                    '50 kHz': ('ENNOID-FRF/Q-V-RL5A-50kHz-9CTB/',
                               'FSESC-420-FRF/Q-V-RL5A-50kHz/',
                               'FSESC-7550-FRF/Q-V-RL5A-50kHz/')}

for frequency in data_directories.keys():
    fig, ax = plt.subplots()
    for data_directory in data_directories[frequency]:
        hardware = data_directory[:data_directory.find('-FRF')]
        title = frequency + ' Plant Measurement'
        vf.plot_plant(data_directory, ax=ax,
                      plot_phase=False, plot_show=False, plot_sim=False, 
                      title=title, mag_label=hardware, alpha=0.5)
```
